Ekstraksi Fitur/ Pre-Processing Data

```python
# # cek nilai unik tiap kolom kategorik
# print(len(df['Location'].unique())) 
# print(len(df['Fuel_Type'].unique())) 
# print(len(df['Transmission'].unique())) 
# print(len(df['Owner_Type'].unique()))

# mengubah nilai kategorik menjadi numerik
# dengan sklearn LabelEncoder
from sklearn import preprocessing

# location
le = preprocessing.LabelEncoder()
le.fit(df['Location'])
df['Location'] = le.transform(df['Location'])

# fuel_type
le = preprocessing.LabelEncoder()
le.fit(df['Fuel_Type'])
df['Fuel_Type'] = le.transform(df['Fuel_Type'])

# transmission
le = preprocessing.LabelEncoder()
le.fit(df['Transmission'])
df['Transmission'] = le.transform(df['Transmission'])

# owner_type
le = preprocessing.LabelEncoder()
le.fit(df['Owner_Type'])
df['Owner_Type'] = le.transform(df['Owner_Type'])
```

```python
df
```

```python
a = "25.44 kmpl" # bentuknya masih string campur antara angka dan satuan
a.split()[0] # fungsi split bisa untuk split string

# buat fungsi untuk sensplit string menjadi pasangan [angka, satuan]
def splitSatuan(text):
  hasilSplit = text.split() #split berdasar spasi
  angka = hasilSplit[0] # akses elemen pertama
  angka = float(angka) # convert tipe string mjd float
  return angka
splitSatuan('26.6 km/kg')

# hapus satuan dari string
df['Mileage'] = df['Mileage'].astype(str)
df['Mileage'] = df['Mileage'].apply(splitSatuan)
df['Engine'] = df['Engine'].astype(str)
df['Engine'] = df['Engine'].apply(splitSatuan)
```

```python
# jadikan satu kolom-kolom yg sudah diproses (yg sudah dikonversi ke angka)
dfBaru = df[['Mileage', 'Engine', 'Location', 'Fuel_Type', 'Transmission', 'Owner_Type', 'Kilometers_Driven','Seats','Year', 'Price']]
dfBaru.head()
```

```python
# mengisi missing value di Mileage dan Engine dengan mean
meanMileage = dfBaru['Mileage'].mean()
meanEngine = dfBaru['Engine'].mean()
dfBaru['Mileage'].fillna(value=meanMileage, inplace=True)
dfBaru['Engine'].fillna(value=meanEngine, inplace=True)
```

```python
# mencek ulang missing value, setelah dilakukan fillna
for item in ['Mileage', 'Engine', 'Location', 'Fuel_Type', 'Transmission', 'Owner_Type', 'Kilometers_Driven','Seats','Year', 'Price']:
  print(item)
  print(dfBaru[item].isnull().sum())
```

Membagi data/mesplit data kedalam 3 bagian yaitu train,val,dan test

```python
from sklearn.model_selection import train_test_split
```

```python
# membagi data train dan test dengan 80% data train dan 20% data test

X = df[['Name','Location', 'Year','Kilometers_Driven', 'Fuel_Type', 'Transmission', 'Owner_Type', 'Mileage', 'Engine', 'Power', 'Seats','Price']]
y = df[['Price']]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# membagi data train dan val , dengan data val adalah 20% dari data train 80%

X_train, X_val, y_train, y_val = train_test_split(X_train,y_train, train_size=0.8) 
```

```python
# mengecek untuk memastikan jumlah data benar

print(X_train.shape), print(y_train.shape)
print(X_val.shape), print(y_val.shape)
print(X_test.shape), print(y_test.shape)
```

membuat model dengan data train

```python
from sklearn import ensemble
# memodelkan
reg = ensemble.GradientBoostingRegressor() # memanggil model awal, dengan default parameter
reg.fit(X_train, y_train) # melatih dengan data train
```

```python
# memprediksi harga bekas untuk data val
ypred = reg.predict(X_val)
ypred
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
# melihat penampakan harga asli data y_val
ytrue = y_val
ytrue
```

```python
# menghitung peforma dengan mengecek nilai error

mean_absolute_error(ytrue, ypred)
```

Memprediksi harga bekas mobil dengan data testing

```python
# mempridiksi harga bekas dengan data testing
ypredict = reg.predict(X_test)
ypredict
```

```python
# melihat penampakan harga asli data y_test
yactual = y_test
yactual
```

```python
# mengecek peforma dengan melihat nilai error dari data test

mean_absolute_error(ypredict, yactual)
```
